# Imports

```python
import glob

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Read data

```python
files = sorted(glob.glob("*.csv"))

df = pd.DataFrame()

for fpath in files:
    df_ = pd.read_csv(fpath)
    if df.size == 0:
        df = df_.copy()
    else:
        df = pd.concat([df, df_])
        
print(df.shape)
```

```python
name_dict = {
    'similarity_sequential': 'Sequential',
    'similarity_shared': 'CUDA shared memory',
    'similarity_global': 'CUDA global memory'
}
df['sim_method'] = df['sim_method'].map(name_dict)
```

# Analysis

## Time vs method used and number of threads

```python
dims = 128
df_ = df[(df['dims'] == dims)].copy()

fig, ax = plt.subplots(figsize=(8,6))
pivot = df_.pivot_table(index=['nthreads'], columns='sim_method', values='time')
pivot.plot(ax=ax)
# plt.xscale('log')
plt.yscale('log')
plt.xticks(size=14)
plt.yticks(size=14)
plt.xlabel('Number of threads', size=16)
plt.ylabel('log time, s', size=16)
plt.title(f'Different methods for size {dims}', size=18);
```
